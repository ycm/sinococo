```python
from database import *
from collections import Counter, defaultdict
```

```python
db = load_database()
```

```python
parse_to_count = Counter()
```

```python
for char, parselist in db.items():
    for parse in parselist:
        parse_to_count[parse[0]] += 1
```

```python
chars_without_distinguishing_parse = set()
for char, parselist in db.items():
    if all(parse_to_count[parse[0]] > 1 for parse in parselist):
        chars_without_distinguishing_parse.add(char)
```

```python
parses_string_to_chars = defaultdict(list)
for c in list(chars_without_distinguishing_parse):
    parses = ''.join(sorted([parse[0] for parse in db[c]]))
    parses_string_to_chars[parses].append(c)
```
