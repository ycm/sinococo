```python
char_to_alts = defaultdict(list)
```

```python
curr = 0x2f00
for radical in KX_chars.split('\n'):
    char = radical[0]
    if radical[0] == '"':
        radical = radical[1:-1]
        char = radical[0]
        alts = radical[2:-1].split('、')
        char_to_alts[char] = alts
    char_to_alts[char].append(chr(curr))
    curr += 1
```

```python
with open('char_to_alts.json', 'w', encoding='utf-8') as f:
    json.dump(char_to_alts, f, indent=2, ensure_ascii=False)
```
