```python
from collections import Counter
```

```python
ct = Counter()
def bfs_add(root):
    global ct
    q = [(0, root)]
    while len(q) != 0:
        level, curr = q[0]
        q = q[1:]
        if curr.classification == 'char':
#             print('#' * level, curr.symbol)
            ct[curr.symbol] += 1
            if curr.child:
                q.append((level + 1, curr.child))
        else:
#             print('$' * level, curr.symbol)
            for c in curr.children:
                q.append((level + 1, c))
```

```python
bfs_add(char_to_ids['隆'])
```

```python
for char in char_to_ids:
    print(char)
    bfs_add(char_to_ids[char])
```
